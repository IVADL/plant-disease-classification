```python
from model import ClassificationModel
```

```python
# Get Image Size
import PIL
image = PIL.Image.open("dataset\\test\\Tomato___Tomato_Yellow_Leaf_Curl_Virus\\image (999).JPG")
width, height = image.size
print(image.size)
```

```python
f = open("dataset/label.txt", 'r')
lines = f.readlines()
label = []
for line in lines:
    label.append(line.strip('\n'))
f.close()

print(label)
```

```python
model = ClassificationModel(
            class_list=label,
            img_width=width,
            img_height=height,
            batch_size=16,
        )
```

```python
model.generate_train_val_data(data_dir="dataset/train/")
```

```python
status = model.set_model(model_name="mobilenet_v2")
if status:
    print('ready model.')
```

```python
history = model.train(epochs=10)
```

```python
model.show_accuracy()
```

```python
model.show_loss()
```

```python
model.evaluation(batch_size=16, data_dir="dataset/test/")
```
